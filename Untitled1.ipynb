Detección de comundiades, Caso práctico

Importar las librerias 

```python
import pandas as pd
import warnings
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import networkx as nx
import networkx.algorithms.community as nx_comm
warnings.filterwarnings("ignore")
```

Lectura de datos de un ejemplo de facebook

```python
facebook = pd.read_csv(
    "https://raw.githubusercontent.com/JoaquinAmatRodrigo/Estadistica-machine-learning-python/master/data/facebook_combined.txt",
    header = None,
    sep    = " ",
    names  = ["user_1", "user_2"],
)
```

```python
facebook = facebook[:5000]
facebook.head()
```

Grafo unidireccional, representa las relaciones de amistad en las redes sociales 

```python
G_facebook = nx.from_pandas_edgelist(
                df           = facebook,
                source       = "user_1",
                target       = "user_2",
                create_using = nx.Graph()
             )
```

Información sobre número de nodos y enlaces presentes en el grafo

```python
print("Número de nodos:", G_facebook.number_of_nodes())
print("Número de enlaces:", G_facebook.number_of_edges())
```

Visualización de la red, se grafica a traves de matplotlib

```python
fig, ax = plt.subplots(figsize=(9, 5))
spring_pos = nx.spring_layout(G_facebook)
nx.draw_networkx(
    G           = G_facebook,
    pos         = spring_pos,
    with_labels = False,
    node_size   = 15,
    ax          = ax
)
```

Detección de comundiades a través de el algoritmo de Louvain

```python
comunidades = nx_comm.louvain_communities(
                G          = G_facebook,
                weight     = 'weight',
                resolution = 0.2
             )
print(f"Número de comunidades detectadas: {len(comunidades)}")
print(f"Tamaño de las comunidades detectadas: {[len(comunidad) for comunidad in comunidades]}")
```

Se procede a difereciar las comunidades detectadas mediante una distinción de colores

```python
color_pallet = list(mcolors.TABLEAU_COLORS.values())

comunidades_map = {}
for i in range(len(comunidades)):
    comunidades_map.update(dict.fromkeys(comunidades[i], i))

comunidades_map

color_nodos = []
for node in G_facebook:
    color_nodos.append(color_pallet[comunidades_map[node]])

fig, ax = plt.subplots(figsize=(9, 5))
spring_pos = nx.spring_layout(G_facebook)
nx.draw_networkx(
    G           = G_facebook,
    pos         = spring_pos,
    with_labels = False,
    node_size   = 15,
    node_color  = color_nodos,
    ax          = ax
)
```
